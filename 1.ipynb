```python
import numpy as np
import pandas as pd
from pyproj import Geod
import datetime
```

```python
train_data = pd.read_csv( "../input/train.csv" )
```

```python
# Calculating distance (m) based on longitude/latituides and adding it in a new column 'dist'
wgs84_geod = Geod(ellps='WGS84')

#Get distance between pairs of lat-lon points
def get_distance(lat1,lon1,lat2,lon2):
    az12,az21,dist = wgs84_geod.inv(lon1,lat1,lon2,lat2)
    return dist

train_data['dist'] = get_distance(train_data['pickup_latitude'].tolist(), train_data['pickup_longitude'].tolist(),
                                  train_data['dropoff_latitude'].tolist(), train_data['dropoff_longitude'].tolist())


train_data = train_data.drop(['pickup_latitude','pickup_longitude','dropoff_latitude','dropoff_longitude'],axis=1)
```

```python
# Replacing N of store_and_fwd_flag with 0 and Y with 1
train_data.replace({'N': 0, 'Y': 1})

train_data['pickup_datetime'] = pd.to_datetime(train_data['pickup_datetime'])
new_dates, new_times = zip(*[(d.weekday(), d.time()) for d in train_data['pickup_datetime']])
train_data = train_data.assign(pickup_weekday=new_dates, pickup_time=new_times)
```
